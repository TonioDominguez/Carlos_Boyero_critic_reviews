```python
def extract_text(soup, tag, attrs):
    elements = soup.find_all(tag, attrs=attrs)
    texts = [element.get_text() for element in elements]
    return texts

headings_texts_all = []
paragraphs_texts_all = []
dates_texts_all = []

for url in urls:
    response = requests.get(url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        
        headings_texts = extract_text(soup, "h2", {"class": "c_t c_t-i"})
        paragraphs_texts = extract_text(soup, "p", {"class": "c_d"})
        dates_texts = extract_text(soup, "a", {"id": "sc_date"})
        
        headings_texts_all.append(headings_texts)
        paragraphs_texts_all.append(paragraphs_texts)
        dates_texts_all.append(dates_texts)
```

<h3 style="color: #008080;">Checking the correspondence between values</h3>

```python
headings_per_list = []
headings_index_to_erase = []

for element in headings_texts_all:
    num_elements = len(element)
    headings_per_list.append(num_elements)

for i, num_elements in enumerate(headings_per_list):
    if num_elements != 27:
        headings_index_to_erase.append(i)
        print(f"Indice {i} despareja ({num_elements})")

print(f"Un total de {len(headings_index_to_erase)} que no coinciden")        
```

```python
paragraphs_per_list = []
paragraphs_index_to_erase = []

for element in paragraphs_texts_all:
    num_elements = len(element)
    paragraphs_per_list.append(num_elements)

for i, num_elements in enumerate(paragraphs_per_list):
    if num_elements != 27:
        paragraphs_index_to_erase.append(i)
        print(f"Indice {i} despareja ({num_elements})")

print(f"Un total de {len(paragraphs_index_to_erase)} que no coinciden")
```

```python
dates_per_list = []
dates_index_to_erase = []

for element in dates_texts_all:
    num_elements = len(element)
    dates_per_list.append(num_elements)

for i, num_elements in enumerate(dates_per_list):
    if num_elements != 27:
        dates_index_to_erase.append(i)
        print(f"Indice {i} despareja ({num_elements})")

print(f"Un total de {len(dates_index_to_erase)} que no coinciden")
```

<h3 style="color: #008080;">Wrangling the lists to square the dataset</h3>

- Transform lists of texts to DF

```python
headings_df = pd.DataFrame(headings_texts_all).T
```

```python
heading_df = headings_df.fillna("Not provided")
```

```python
heading_df
```

```python
paragraphs_df = pd.DataFrame(paragraphs_texts_all).T
```

```python
paragraphs_df = paragraphs_df.fillna("Not provided")
```

```python
dates_df = pd.DataFrame(dates_texts_all).T
```

```python
dates_df = dates_df.fillna("Not provided")
```

- Melting each DF into 1 variable to join them at the final DF

```python
heading_melt = pd.melt(heading_df)
```

```python
heading_melt.head(1)
```

```python
paragraphs_melt = pd.melt(paragraphs_df)
```

```python
dates_melt = pd.melt(dates_df)
```

<h3 style="color: #008080;">Melted DF</h3>

```python
boyero_df = pd.DataFrame()
```

```python
boyero_df["header"] = heading_melt["value"]
boyero_df["paragraph"] = paragraphs_melt["value"]
boyero_df["date"] = dates_melt["value"]
```

```python
boyero_df.head(1)
```

<h3 style="color: #008080;">Formating "date" column</h3>

```python
boyero_df.dtypes
```

```python
def date_cleaner(date):
    clean_date = date.split("-")[0].strip()
    return clean_date
```

```python
boyero_df["date"] = boyero_df["date"].apply(date_cleaner)
```

```python
boyero_df["date"].head(5)
```

```python
months = {
    "ene": "01",
    "feb": "02",
    "mar": "03",
    "abr": "04",
    "may": "05",
    "jun": "06",
    "jul": "07",
    "ago": "08",
    "sept": "09",
    "oct": "10",
    "nov": "11",
    "dic": "12"
}
```

```python
boyero_df["date"] = boyero_df["date"].apply(lambda x: f"{x.split(' ')[0]} {months[x.split(' ')[1]]} {x.split(' ')[2]}")
```

```python
boyero_df.head(5)
```

```python
boyero_df["date"] = pd.to_datetime(boyero_df["date"], format="%d %m %Y")
```

<h3 style="color: #008080;">Selecting dates from 2020 to 2024</h3>

```python
boyero_dates = boyero_df[boyero_df["date"] >= "2020-01-01"]
```
